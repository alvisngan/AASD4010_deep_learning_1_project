```python
#import loss samples

import numpy as np
loss_samples = np.load('loss_samples.npy')
accuracy_samples = np.load('accuracy_samples.npy')
```

```python
# Calculate mean and variance
accuracy_values = [sample for sample in accuracy_samples]
mean_accuracy = np.mean(accuracy_values)
variance_accuracy = np.var(accuracy_values)

print(f"Mean Accuracy: {mean_accuracy:0.4f}")
print(f"Variance of Accuracy: {variance_accuracy:0.4f}")

x_vals_acc = np.linspace(min(accuracy_values), max(accuracy_values), 1000)
# Calculate the y values using the Gaussian PDF
y_vals_acc = (1 / (np.sqrt(2 * np.pi * variance_accuracy))) * np.exp(-((x_vals_acc - mean_accuracy) ** 2) / (2 * variance_accuracy))

```

```python
#plot histogram of loss samples
import matplotlib.pyplot as plt
plt.hist(loss_samples, bins=100)
plt.xlabel('Loss')
plt.ylabel('Frequency')
plt.title('Loss for random model parameters, 100k-models')
plt.show()
```

```python
# Plot the histogram and best-fit curve
plt.hist(accuracy_samples, bins=40, alpha=0.7, label='Samples', density=True)
plt.plot(x_vals_acc, y_vals_acc, color='red', label=r'$\mu={:.4f}, \sigma^2={:.4f}$'.format(mean_accuracy, variance_accuracy))

#X axis limits
#plt.xlim(0, 1.0)

plt.xlabel('Accuracy')
plt.ylabel('Density')
plt.title('Accuracy Distribution for random model parameters, 100k-models')
plt.legend()
plt.show()
```
